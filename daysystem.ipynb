```python
import os
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt

np.set_printoptions(suppress=True)

def datetime2Seconds(list):
    newlist = []
    for elem in list:
        newlist.append((elem - dt.datetime(2018,3,31)).total_seconds())
    return newlist
```

```python
df = pd.read_excel('./excel/Assignment2_Dataset2018.xlsx', skiprows=1)

day1times = df['Day 1']
d1_sec = datetime2Seconds(day1times)

print(d1_sec)

inittime = d1_sec[0]
d1_secdiff = []
for t in d1_sec[1:]:
    d1_secdiff.append(t - inittime)
    inittime = t

d1_secdiff = np.array(d1_secdiff)
d1_mean = d1_secdiff.mean()
```

```python
print(d1_secdiff)
print("Mean is : " + str(d1_mean))
plt.plot(d1_secdiff)
plt.ylabel('Interarrival Time')
plt.show()
```
